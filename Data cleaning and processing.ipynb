```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
frame=pd.read_csv('end_565001.csv')
```

```python
frame
```

```python
frame=frame.drop_duplicates(inplace=False).dropna().reset_index()
```

```python
frame.info()
```

###### Чистим данные от utm меток

```python
frame['ga:landingPagePath']=frame['ga:landingPagePath'].str.rsplit("?", n=1, expand=True).drop(1,axis=1)
frame
```

```python
frame['ga:pagePath']=frame['ga:pagePath'].str.rsplit("?", n=1, expand=True).drop(1,axis=1)
frame
```

```python
frame['ga:secondPagePath']=frame['ga:secondPagePath'].str.rsplit("?", n=1, expand=True).drop(1,axis=1)
frame['ga:exitPagePath']=frame['ga:exitPagePath'].str.rsplit("?", n=1, expand=True).drop(1,axis=1)
frame['ga:previousPagePath']=frame['ga:previousPagePath'].str.rsplit("?", n=1, expand=True).drop(1,axis=1)
```

```python
frame[(frame['ga:country'] == '(other)')] #удаляем строку с неопределёнными данными
```

```python
frame=frame.drop(labels = [52272], axis = 0)
```

###### Отбираем страны для анализа

```python
country_page=pd.DataFrame(frame.groupby('ga:country')['ga:pageviews'].sum()).sort_values(by='ga:pageviews', ascending=False).reset_index()
country_page
```

```python
cp=country_page.head(11) #sum GA:PAGEVIEWS > 5K
cp
```

```python
country_clients=pd.DataFrame(frame.groupby('ga:country').count()).sort_values(by='ga:landingPagePath', ascending=False).drop("index",axis=1).reset_index()
country_clients
```

```python
country_clients.head(20) #count GA:PAGEVIEWS > 2K
```
