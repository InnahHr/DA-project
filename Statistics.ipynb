```python
import pandas as pd
import numpy as np
```

```python
import os
import glob
```

```python
path = r'C:\Users\inna_\OneDrive\Рабочий стол\DA2021\выгрузки'
all_files = glob.glob(os.path.join(path , "query-explorer-export*.tsv"))
```

```python
li = []
```

```python
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, sep="\t")
    li.append(df)
```

```python
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame=frame.drop_duplicates(inplace=False)
```
